```python
def buscaEmLargura(noPartida,noChegada,evitar = None):
  fila = []
  fila.append(noPartida)
  nosVisitados = [0, 0, 0, 0, 0, 0, 0, 0, 0]
  if evitar is not None:
    for nope in evitar:
      nosVisitados[nope-1]=1
  caminho = dict()
  while len(fila)>0:
    no = fila.pop(0)
    nosVisitados[no-1] = 1
    if no == noChegada:
      break
    possiveis= []
    for n in grafo[no-1]:
      if nosVisitados[n-1] == 0:
        nosVisitados[n-1] = 1
        fila.append(n)
        possiveis.append(n)
        if n == noChegada:
          fila = list()
          break
    caminho[no] = possiveis
    
  final = list()
  res = dict(reversed(list(caminho.items())))
  number = noChegada
  caminho = list()
  caminho.append(number)
  for i in range(len(res)):
    if number in res[list(res.keys())[i]]:
      number = list(res.keys())[i]
      caminho.append(number)
  return list(reversed(caminho))
```
